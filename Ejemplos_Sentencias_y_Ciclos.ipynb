# 📚 Ejemplos de Sentencias Condicionales y Ciclos en Python
Explicaciones básicas y ejemplos prácticos.

```python
# Ejemplo 1: Sentencias condicionales (if, elif, else)
edad = 20
if edad < 13:
    print("Eres un niño")
elif edad < 18:
    print("Eres un adolescente")
else:
    print("Eres un adulto")
```

```python
# Ejemplo 2: Bucle for (recorriendo una lista)
frutas = ["manzana", "plátano", "cereza"]
for fruta in frutas:
    print("Me gusta comer", fruta)
```

```python
# Ejemplo 3: Uso de range() en un bucle for
print("\nUso de range() para imprimir números del 0 al 4:")
for numero in range(5):
    print("Número:", numero)
```

```python
# Ejemplo 4: Bucle while (contador sencillo)
contador = 0
while contador < 5:
    print("Contador:", contador)
    contador += 1
```

```python
# Ejemplo 5: Uso de break (salir de un bucle antes de que termine)
for numero in range(10):
    if numero == 5:
        print("Se detiene el bucle en el número 5")
        break
    print("Número:", numero)
```

```python
# Ejemplo 6: Uso de continue (saltar una iteración)
for numero in range(5):
    if numero == 2:
        continue
    print("Número:", numero)
```

```python
# Ejemplo 7: Uso de pass (placeholder)
for letra in "Python":
    if letra == "h":
        pass
    else:
        print("Letra actual:", letra)
```
