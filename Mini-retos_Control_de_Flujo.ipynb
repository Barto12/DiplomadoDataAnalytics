# 🧠 Mini-retos de Control de Flujo en Python
Ejercicios para practicar el uso de `range()`, `for`, `if`, `break` y `continue`.

```python
# 🏁 Reto 1: Imprimir solo números pares del 0 al 20
for num in range(0, 21):
    if num % 2 == 0:
        print(num)
```

```python
# ⏳ Reto 2: Cuenta regresiva del 10 al 1
for num in range(10, 0, -1):
    print(num)
```

```python
# ➡️ Reto 3: Crear una lista de múltiplos de 3 entre 0 y 30
multiplos_3 = []

for num in range(0, 31):
    if num % 3 == 0:
        multiplos_3.append(num)

print(multiplos_3)
```

```python
# 🛑 Reto 4: Imprimir los primeros 5 números pero detenerse si encuentra un número mayor que 3
for num in range(5):
    if num > 3:
        break
    print(num)
```

```python
# 🔁 Reto 5: Saltar el número 4 usando `continue`
for num in range(6):
    if num == 4:
        continue
    print(num)
```
